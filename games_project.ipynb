# Анализ данных о компьютерных играх

## Описание исследования

В распоряжении имеются данные о компьютерных играх: название игры, платформа, год выпуска, жанр, продажи в разных регионах, оценки критиков и пользователей, а также возрастной рейтинг.

Эти данные позволяют изучить структуру игрового рынка, выявить особенности игр, выпущенных в разные годы и на разных платформах, а также подготовить данные для дальнейшего анализа.

## Цель исследования

Цель исследования — подготовить данные о компьютерных играх для дальнейшего анализа и изучить основные характеристики игр за актуальный период.

В рамках работы необходимо обработать пропуски, исправить ошибки в данных, удалить дубликаты, выделить актуальный период и определить наиболее представленные платформы по количеству выпущенных игр.

## Задачи исследования

Для достижения цели необходимо:

1. Изучить общую информацию о датасете.
2. Привести названия столбцов к единому стилю.

5. Проверить данные на наличие явных и неявных дубликатов.

7. Выделить актуальный период для дальнейшего анализа.
8. Провести категоризацию оценок пользователей и критиков.
9. Определить топ-7 платформ по количеству выпущенных игр за актуальный период.
10. Сформулировать общий вывод по результатам подготовки и первичного анализа данных.

## Исходные данные

В исследовании используется таблица с данными о компьютерных играх.

После обработки данных пропуски остались в основном в столбцах critic_score и user_score: около 51.39% и 40.13% соответственно. Эти значения не заполнялись, так как оценки критиков и пользователей невозможно корректно восстановить по имеющимся данным без искажения анализа.
В year_of_release осталось около 1.62% пропусков. В столбцах name, genre, eu_sales, jp_sales доля пропусков незначительная.
Пропуски в rating были заменены на значение unknown, поэтому после обработки в этом столбце отсутствуют значения NaN.

Наибольшая доля пропусков наблюдается в столбцах critic_score и user_score. 
Пропуски в user_score могут быть связаны с тем, что не все пользователи оставляют оценки играм, особенно если игра была малопопулярной, старой или выходила на менее массовой платформе.

Пропуски в critic_score, вероятно, объясняются тем, что профессиональные критики оценивают не все игры. Больше внимания обычно получают крупные и заметные релизы, тогда как нишевые, старые или региональные проекты могли не иметь достаточного количества рецензий. Поэтому отсутствие оценки критиков не стоит заменять средним или медианным значением, так как это исказило бы реальные данные.

В столбцах с региональными продажами были обнаружены единичные пропуски. Их доля крайне мала относительно общего объёма датасета, поэтому было принято решение не удалять эти строки. Заполнять такие значения средним или медианным значением некорректно, так как продажи игр распределены неравномерно: небольшое количество популярных игр может иметь очень высокие продажи и заметно искажать средние показатели. Поэтому пропуски в столбцах с продажами были оставлены без изменения, чтобы не вносить искусственные значения в данные.

```python
def categorize_user_score(score):
    if pd.isna(score):
        return 'без рейтинга'
    elif 8 <= score <= 10:
        return 'высокая оценка'
    elif 3 <= score < 8:
        return 'средняя оценка'
    elif 0 <= score < 3:
        return 'низкая оценка'
    else:
        return 'некорректное значение'
```

```python
def categorize_critic_score(score):
    if pd.isna(score):
        return 'без рейтинга'
    elif 80 <= score <= 100:
        return 'высокая оценка'
    elif 30 <= score < 80:
        return 'средняя оценка'
    elif 0 <= score < 30:
        return 'низкая оценка'
    else:
        return 'некорректное значение'
```

```python
df_actual['user_score'] = pd.to_numeric(df_actual['user_score'], errors='coerce')
```

```python
df_actual['categorize_user_score'] = df_actual['user_score'].apply(categorize_user_score)
```

```python
df_actual['critic_score'] = pd.to_numeric(df_actual['critic_score'], errors='coerce')
```

```python
df_actual['critic_score_category'] = df_actual['critic_score'].apply(categorize_critic_score)
```

```python
top_platforms = df_actual['platform'].value_counts().head(7)
```

```python
print(top_platforms)
```

Для анализа был выделен актуальный период с 2000 по 2013 год включительно. На его основе был создан отдельный датафрейм `df_actual`, чтобы дальнейшие расчёты проводились только по играм, выпущенным в выбранный период.

Также были созданы пользовательские функции для категоризации оценок пользователей и критиков. Для `user_score` использовалась шкала от 0 до 10, а для `critic_score` — шкала от 0 до 100. Оценки были разделены на категории: низкая, средняя, высокая, а пропущенные значения были вынесены в отдельную категорию «без рейтинга». Это позволяет анализировать оценки не только как числовые значения, но и как более удобные смысловые группы.

После этого были выделены топ-7 платформ по количеству выпущенных игр за актуальный период. Самыми представленными платформами оказались PS2, DS, Wii, PSP, X360, PS3 и GBA. Больше всего игр было выпущено на PS2 — 2154 игры, почти столько же на DS — 2146 игр. Далее идут Wii — 1294 игры, PSP — 1199 игр, X360 — 1138 игр, PS3 — 1107 игр и GBA — 826 игр.

Таким образом, в актуальном периоде наиболее популярными по количеству выпущенных игр стали платформы PS2 и DS. Это говорит о том, что именно они занимали наиболее заметное место в датасете по числу релизов за выбранные годы.
